### 청년층/노년층 , 동양/서양별 이미지 저장

```python
import pandas as pd

df = pd.read_csv('./data/age_gender.csv')
df
```

### 연령층 나누기
### age_group : 0~49세 == 0 (청년층) , 50세 == 1 (노년층)

```python
df.loc[(df['age']>=0) & (df['age']<50),'age_group'] = 0
df.loc[df['age']>=50,'age_group'] = 1
df
```

```python
# float => int

df['age_group'] = df['age_group'].astype('int')
```

```python
df['age_group'].value_counts()
```

### 인종 나누기
### ethnicity : 동양 == 0, 서양 == 1

```python
df.loc[df['ethnicity'] == 2,'ethnicity_group'] = 0
df
```

```python
df['ethnicity_group'].value_counts()
```

```python
df.loc[df['ethnicity_group'].isna(),'ethnicity_group'] = 1.0
df
```

```python
df['ethnicity_group'] = df['ethnicity_group'].astype('int')
df
```
