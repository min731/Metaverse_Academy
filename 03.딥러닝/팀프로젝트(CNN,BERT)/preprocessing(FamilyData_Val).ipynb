### 연령/각도별 안면 데이터 저장

```python
from glob import glob
import os
root_path = 'E:/045.가족 관계가 알려진 얼굴 이미지 데이터/01.데이터/2.Validation/원천데이터/'

VS_path = os.listdir(root_path)
print(len(VS_path))
VS_path 
```

```python
tmp = []
for VS in VS_path:
    if VS.find(".zip")==-1:
        tmp.append(VS)
print(len(tmp))
VS_path = tmp
VS_path
```

```python
import shutil
from tqdm import tqdm

# root_path = 'E:/045.가족 관계가 알려진 얼굴 이미지 데이터/01.데이터/2.Validation/원천데이터/'
save_path = 'E:/045.가족 관계가 알려진 얼굴 이미지 데이터/01.데이터/Preprocessed_Datasets/'

for VS in tqdm(VS_path):
    print("TS")
    AB_path = os.listdir(root_path+VS+'/')
    if len(os.listdir(root_path+VS+'/'+'A(친가)'+'/2.Individuals'))!=0:
        image_path = glob(root_path+VS+'/'+'A(친가)'+'/2.Individuals/*.jpg')
        for image in image_path:
            if image.split("\\")[-1].find('CAM') == -1:
                shutil.copyfile(image,save_path+image.split("\\")[-1]) 
    else:
        image_path = glob(root_path+VS+'/'+'B(외가)'+'/2.Individuals/*.jpg')
        for image in image_path:
            
            if image.split("\\")[-1].find('CAM') == -1:
                shutil.copyfile(image,save_path+image.split("\\")[-1]) 
```

### Multi Label Classification을 위한 DataFrame 화

```python
import pandas as pd

df = pd.DataFrame(columns=['img_path','M','F',"0's","10's","20's","30's","40's","50's","60+"])
df
```

```python
save_path = 'E:/045.가족 관계가 알려진 얼굴 이미지 데이터/01.데이터/Preprocessed_Datasets/'

datasets = glob(save_path+'*.jpg')
datasets[:5]
```

```python
img_paths = []
for data in datasets:
    img_path = data.split("\\")[-1]
    img_paths.append(img_path)
# img_paths

df['img_path'] = img_paths
df
```

```python
# columns

# img_paths = []
# gender_M = []
# gender_F = []
# age_20 = []
# age_30 = []
# age_40 = []
# age_50 = []
# age_60 = []

gender_M_list = ['GF','F','S','S1','S2','S3','S4']
gender_F_list = ['GM','M','D','D1','D2','D3','D4']
```

```python
for idx,data in enumerate(datasets):
    label = data.split("\\")[-1]
    img_paths.append(label)

    gender = label.split('_')[2]
    age = int(label.split('_')[3])
    print(age)

    if gender in gender_M_list:
        df.loc[idx,'M'] = 1
    else:
        df.loc[idx,'F'] = 1

    if age>=0 and age<10:
        df.loc[idx,"0's"] = 1
    elif age>=10 and age<20:
        df.loc[idx,"10's"] = 1
    elif age>=20 and age<30:
        df.loc[idx,"20's"] = 1
    elif age>=30 and age<40:
        df.loc[idx,"30's"] = 1
    elif age>=40 and age<50:
        df.loc[idx,"40's"] = 1
    elif age>=50 and age<60:
        df.loc[idx,"50's"] = 1
    elif age>=60:
        df.loc[idx,"60+"] = 1
```

```python
df
```

```python
df = df.fillna(0)
df
```
