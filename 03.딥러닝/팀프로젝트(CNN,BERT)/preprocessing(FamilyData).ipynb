### 연령/각도별 안면 데이터 저장

```python
from glob import glob
import os
root_path = 'C:/Users/ypd04/OneDrive/바탕 화면/045.가족 관계가 알려진 얼굴 이미지 데이터/01.데이터/1.Training/원천데이터_0903_add/TS0001to0020/'

TS_path = os.listdir(root_path)
TS_path
```

```python
import shutil
from tqdm import tqdm

save_path = 'C:/Users/ypd04/OneDrive/바탕 화면/FamilyDatasets/'

for TS in tqdm(TS_path):
    print("TS")
    AB_path = os.listdir(root_path+TS+'/')
    if len(os.listdir(root_path+TS+'/'+'A(친가)'+'/2.Individuals'))!=0:
        image_path = glob(root_path+TS+'/'+'A(친가)'+'/2.Individuals/*.jpg')
        for image in image_path:
            if image.split("\\")[-1].find('CAM') == -1:
                shutil.copyfile(image,save_path+image.split("\\")[-1]) 
    else:
        image_path = glob(root_path+TS+'/'+'B(외가)'+'/2.Individuals/*.jpg')
        for image in image_path:
            
            if image.split("\\")[-1].find('CAM') == -1:
                shutil.copyfile(image,save_path+image.split("\\")[-1]) 
```

### Multi Label Classification을 위한 DataFrame 화

```python
import pandas as pd

df = pd.DataFrame(columns=['img_path','M','F',"10's","20's","30's","40's","50's","60+"])
df
```

```python
save_path = 'C:/Users/ypd04/OneDrive/바탕 화면/FamilyDatasets/'

datasets = glob(save_path+'*.jpg')
datasets[:5]
```

```python
img_paths = []
for data in datasets:
    img_path = data.split("\\")[-1]
    img_paths.append(img_path)
# img_paths

df['img_path'] = img_paths
df
```

```python
# columns

img_paths = []
gender_M = []
gender_F = []
age_20 = []
age_30 = []
age_40 = []
age_50 = []
age_60 = []

gender_M_list = ['GF','F','S']
gender_F_list = ['GM','M','D']
```

```python
for idx,data in enumerate(datasets):
    label = data.split("\\")[-1]
    img_paths.append(label)

    gender = label.split('_')[2]
    age = int(label.split('_')[3])
    print(age)

    if gender in gender_M_list:
        df.loc[idx,'M'] = 1
    else:
        df.loc[idx,'F'] = 1

    if age>=10 and age<20:
        df.loc[idx,"10's"] = 1
    elif age>=20 and age<30:
        df.loc[idx,"20's"] = 1
    elif age>=30 and age<40:
        df.loc[idx,"30's"] = 1
    elif age>=40 and age<50:
        df.loc[idx,"40's"] = 1
    elif age>=50 and age<60:
        df.loc[idx,"50's"] = 1
    elif age>=60:
        df.loc[idx,"60+"] = 1
```
