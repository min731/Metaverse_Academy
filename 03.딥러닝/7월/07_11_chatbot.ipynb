```python
import sentence_transformers
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import random
import numpy as np

# 소숫점 6자리 까지만
np.set_printoptions(precision=6,suppress=True)

# 한국어 모델
model = SentenceTransformer('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
```

```python
sentence = ['안녕하세요 반갑습니다','임베딩이 잘될까요?']
embedding = model.encode(sentence)
embedding
```

### 데이터 로드

```python
df = pd.read_csv('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/ChatbotData.csv')
df
```

```python
import csv

with open('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/ChatbotData_embedding.csv','w') as f:
  mapdata = []
  write = csv.writer(f)

  for tmp in df['Q']:
    emb = model.encode(tmp)
    mapdata.append(emb)

  # 각 셀마다 1차원씩 저장
  write.writerows(mapdata)
```

```python
embedding_data = pd.read_csv('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/ChatbotData_embedding.csv',header=None)
embedding_data
```

```python
def chatbot_text(text):
  em_result = model.encode(text)
  cos_result = []
  for tmp in range(len(embedding_data)):
    data = embedding_data.iloc[tmp]
    cos_result.append(cosine_similarity([data],[em_result])[0][0])

  df['cos'] = cos_result
  df_result = df.sort_values('cos',ascending=False)
  r = random.randint(0,5)

  return df_result.iloc[r]

chatbot_text("오늘 너무 힘들었어")
```

```python
chatbot_text("오늘 점심 뭐먹을까?")
```

```python
import time
start = time.time()
chatbot_text("바다 가고 싶다!")
end = time.time()
print(end-start)
```
