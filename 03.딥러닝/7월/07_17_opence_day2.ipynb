### 그리기 함수

```python
img = np.zeros((400,400,3),dtype=np.uint8) +255
# 선그리기
# CV2 그때그때 마다 RGB or BGR 등 다름
cv2.line(img,(0,0),(200,200),(0,0,255),3,lineType=cv2.LINE_8) # 시작 포인트,끝 포인트, 컬러(BGR), 두께, 디테일
cv2_imshow(img)
```

```python
img = np.zeros((400,400,3),dtype=np.uint8)
cv2.rectangle(img,(10,10),(40,40),(255,0,0),1) # 시작위치, 끝위치, 색상, 두께
cv2.rectangle(img,(200,200,40,80),(0,0,255),2) # 시작위치, 크기, 색상,두께
cv2.circle(img,(200,200),30,(0,0,255),5) # 시작위치, 반지름, 색상, 두께
cv2.circle(img,(200,200),30,(0,0,255),5,lineType=cv2.LINE_AA) # 시작위치, 반지름, 색상, 두께, 라인타입
cv2_imshow(img)
```

```python
img = np.zeros((400,400,3),dtype=np.uint8)
# 한글을 쓰고 싶으면 PIL로

from PIL import ImageFont,ImageDraw,Image

img = np.zeros((400,400,3),dtype=np.uint8)
img = Image.fromarray(img) # img 배열을 PIL이 처리가능하게 변환
draw = ImageDraw.Draw(img)
font= ImageFont.truetype('NanumGothic.ttf',20)
# 글씨 쓰는 위치
org = (100,100)
text = "이제 한글로 적을 수 있다."
draw.text(org,text,font=font,fill=(255,255,255))
img = np.array(img)

cv2.putText(img,'Test테스트',(50,50),cv2.FONT_HERSHEY_DUPLEX,2,(0,255,0)) # 텍스트,위치,폰트,크기,색상
cv2_imshow(img)
```
