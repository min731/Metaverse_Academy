### GrayScale 이미지 Blur 처리

```python
import numpy as np

src = cv2.imread('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/gerbera jamesonii.jpg',cv2.IMREAD_GRAYSCALE)
src_f = src.astype(np.float32)
blr = cv2.GaussianBlur(src_f,(0,0),6.0)

dst = np.clip(2.*src_f-blr,0,255).astype(np.uint8)
```

```python
cv2_imshow(dst)
```

```python
src = cv2.imread('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/gerbera jamesonii.jpg')
src_ycrcb = cv2.cvtColor(src,cv2.COLOR_BGR2YCrCb)

src_f = src_ycrcb[:,:,0].astype(np.float32)
blr = cv2.GaussianBlur(src_f,(0,0),6.0)
src_ycrcb[:,:,0] = np.clip(2.*src_f-blr,0,255).astype(np.uint8)
dst = cv2.cvtColor(src_ycrcb,cv2.COLOR_YCrCb2BGR)
cv2_imshow(dst)
```
