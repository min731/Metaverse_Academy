# 어파인/투시 변환 : OCR에서 많이 사용

```python
import cv2
from google.colab.patches import cv2_imshow # 코랩에서는 cv2.imshow()하면 안됌

src = cv2.imread('/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/7월/data/business-card_640.jpg')
cv2_imshow(src)
```

```python
import numpy as np

w,h = 720,480

# 대응되는 좌표값을 정확하게 맞추어야함
srcQuad = np.array([[160,207],[395,112],[487,221],[239,327]],np.float32)
dstQuad = np.array([[0,0],[w-1,0],[w-1,h-1],[0,h-1]],np.float32)

pers = cv2.getPerspectiveTransform(srcQuad,dstQuad)
dst = cv2.warpPerspective(src,pers,(w,h))

cv2_imshow(dst)
```
