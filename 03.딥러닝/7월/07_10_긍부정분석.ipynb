```python
import re
from tqdm import tqdm

noun_result = []  #[ [] , [], [] ]

for temp in tqdm(df['text']):
    han = re.compile('[^ã„±-ã…£ê°€-í£]')
    temp_result = han.sub(' ',temp) #ë¬¸ì¥ì—ì„œ í•œê¸€ë§Œ ë‚¨ê¸°ê³  ì œê±°
    temp_noun = okt.nouns(temp_result)
    result_noun = [ x for x in temp_noun if len(x) >1]
    noun_result.append(result_noun)

print(noun_result)
```

```python
# bag of word ë²¡í„° ìƒì„±
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(tokenizer=lambda x:x,lowercase=False)
bow_vect = count_vectorizer.fit_transform(noun_result)
word_list = count_vectorizer.get_feature_names_out()
print(word_list)
print(len(word_list))
```

```python
# if-idf ìƒì„±
from sklearn.feature_extraction.text import TfidfTransformer

tf_vectorizer = TfidfTransformer()
# bow vectorë¥¼ tf_vectorizerì— ë„£ìœ¼ë©´ ëŒ
tf_idf = tf_vectorizer.fit_transform(bow_vect)
# ë‹¨ì–´ì˜ ê°€ì¤‘ì¹˜ê°€ ê³„ì‚°ëŒ
tf_idf.toarray()
```

```python
df['score_real'] = np.where(df['score']>= 4,1,0)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split

x = tf_idf
y = df['score_real']

# ë°ì´í„° ë¶ˆê· í˜• => stratify
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1,stratify=df['score_real'])
```

```python
X_train.shape , X_test.shape
```

```python
# ëª¨ë¸ í•™ìŠµ (ì´ì§„ ë¶„ë¥˜)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

lr_model = LogisticRegression(random_state=1)
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
def comment_predict(text):
  sentence_test = text
  # han = re.compile('[^ã„±- | ê°€-í£]')
  han = re.compile('[^ã„±-ã…£ ê°€-í£]')
  temp_result = han.sub(' ',sentence_test)
  temp_noun = okt.nouns(temp_result)
  result_noun = [x for x in temp_noun if len(x)>1]

  # transformë§Œ (fití•˜ë©´ ì•ˆëŒ)
  vec_test = count_vectorizer.transform([result_noun])
  vec_test.toarray().sum()
  tf_test = tf_vectorizer.transform(vec_test)
  tf_test.toarray().sum()

  result = ''
  if lr_model.predict(tf_test) == 1:
    result = 'ê¸ì •'
  else:
    result = 'ë¶€ì •'

  return result

print(comment_predict('ë©”ë‰´ê°€ ë‹¤ì–‘í•´ì„œ ì¢‹ì•˜ì–´ìš”'))
```
