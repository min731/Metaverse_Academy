```python
import torch
import torch.nn as nn
```

```python
from torchvision.models.vgg import vgg16 # 16개의 레이어
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
# pretrained = True (사전 학습된 가중치 사용)
# pretrained = False
model = vgg16(pretrained=False)
```

```python
#  (avgpool): AdaptiveAvgPool2d(output_size=(7, 7))
#   (classifier): Sequential(
#     (0): Linear(in_features=25088, out_features=4096, bias=True)
#     (1): ReLU(inplace=True)
#     (2): Dropout(p=0.5, inplace=False)
#     (3): Linear(in_features=4096, out_features=4096, bias=True)
#     (4): ReLU(inplace=True)
#     (5): Dropout(p=0.5, inplace=False)
#     (6): Linear(in_features=4096, out_features=1000, bias=True)
#   )
# )
```

```python
fc = nn.Sequential(
    nn.Linear(25088,4096),
    nn.ReLU(),
    # Dropout default p=0.5
    nn.Dropout(),
    nn.ReLU(),
    nn.Linear(4096,4096),
    nn.ReLU(),
    nn.Dropout(),
    nn.Linear(4096,10)
)

# 분류기 바꿔치기함
model.classifier = fc
```

```python
model
```

```python
model.to(device)
```

```python
from torchvision.datasets.cifar import CIFAR10
from torchvision.transforms import Compose,ToTensor,Resize,RandomHorizontalFlip,Normalize
from torch.utils.data.dataloader import DataLoader
from torch.optim.adam import Adam
import torchvision.transforms as T
```

```python
from torchvision.transforms.transforms import RandomCrop
# 사전 지식 : VGG16은 224로 넣어야함

transforms = T.Compose(

    [
     Resize(224),
     RandomCrop((224,224),padding=4),
     RandomHorizontalFlip(),
     ToTensor(),
     Normalize(mean=(0.4914,0.4822,0.4465),std=(0.247,0.243,0.261))

    ]

)
```

```python
train_data = CIFAR10(root='/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/6월/vgg_cifar10/train/',download=True,train=True,transform=transforms)
test_data = CIFAR10(root='/content/drive/MyDrive/메타버스_아카데미_2기/딥러닝/6월/vgg_cifar10/test/',download=True,train=False,transform=transforms)
```

```python
train_loader = DataLoader(train_data,batch_size=32,shuffle=True)
test_loader = DataLoader(test_data,batch_size=32,shuffle=False)
```

```python
from tqdm import tqdm

lr = 1e-4
optim = Adam(model.parameters(),lr=lr)

for epoch in range(10):
  for data,label in tqdm(train_loader):
    optim.zero_grad()
    pred = model(data.to(device))

    loss = nn.CrossEntropyLoss()(pred,label.to(device))
    loss.backward()

    optim.step()

  print(f'epoch : {epoch+1}, loss : {loss.item()} ')
```

```python
model.eval()

total_corr = 0

with torch.no_grad():
  for data, label in test_loader:
    optim.zero_grad()
    pred = model(data.to(device))
    preds = pred.data.max(1)[1]
    corr = preds.eq(label.to(device)).sum().item()

    total_corr += corr
  print(total_corr/len(test_data))
```
