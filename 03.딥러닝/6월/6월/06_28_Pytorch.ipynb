# 딥러닝으로 구현

```python
# 클래스로 만드는 것이 일반적

# nn에 모델 포함됌
import torch.nn as nn
# from torch import nn

# nn.Model 상속받음
class LRModel(nn.Module):

  # 생성자
  # self : 복사본의 정보를 담고 있다
  def __init__(self):
    super(LRModel,self).__init__()

    # tf와 순서다 다름

    # nn.Linear(input,output)
    # 위 데이터에서 피처1개, 결과1개
    self.layer = nn.Linear(1,1)

  # 실제 데이터는 forward로 들어옴
  # x라는 데이터가 들어옴
  def forward(self,x):

    # 위에서 만든 layer에 x값을 집어넣음
    x = self.layer(x)

    return x

# 위 신경망에서는
# input -------------- output
#          w1 , b1
#  y = w1x + b1
```

```python
model = LRModel()

w,b = model.parameters()
w, b
```

```python
import numpy as np

# 가중치의 0번째 값
w1 = w[0][0].item()
# bias는 1차원
b1 = b[0].item()

w1 ,b1
```

```python
x1 = np.array([-30,30])
x1
```

```python
y1 = w1*x1 + b1
y1
```

```python
plt.scatter(x,y)
plt.plot(x1,y1,'r')
plt.show()
```

```python
# 학습시킬 때 필요한 것
# 손실함수, 최적화함수, metrics, epochs, learing rate
# 에폭 돌릴 때마다 최적화 함수를 zero grade로 초기화 시킴
# 그 후 모델에 데이터를 넣고 오차를 구함
# 이를 역전파로 보내어 가중치를 수정함
```

# 파이토치 기본 패턴

```python
# 최적화함수
import torch.optim as optim
# 모델의 파라미터를 최적화해라
optimizer = optim.SGD(model.parameters(),lr=0.001)

# 손실함수는 nn 안에 들어가 있음
# criterion = nn.MSELoss()(예측값,정답값)
criterion = nn.MSELoss()

# 에폭
epochs = 20

# 손실함수 저장
cri_list = []

for epoch in range(epochs):
  # optimizer 0 grade로 초기화
  optimizer.zero_grad()

  # 모델에 학습 데이터 주입
  pred = model(x)

  # 오차값
  loss = criterion(pred,y)

  # 오차값 저장
  cri_list.append(loss.item())

  # 역전파
  loss.backward()

  print(f'{epoch+1} loss : {loss.item()}')

  # 최적화함수로 다음 스텝
  optimizer.step()
```

```python
plt.plot(range(epochs),cri_list)
plt.show()
```
