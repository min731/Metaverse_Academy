# 도로 상황 분류 프로젝트

### 데이터 개요
* 총 30000개 데이터, Normal(맑음) , Rainy(비), Snowy(눈) 3개 클래스
* 전체 데이터 : Youtube 영상->이미지 데이터(15000개) - Normal(1개영상, 5000개), Rainy(1개영상, 5000개), Snowy(1+2개영상, 10000개)
* AIhub 이미지 데이터 (10000개) - Normal(5000개) , Rainy(5000개)

### 데이터 전처리1 : Youtube 원본 영상 데이터 -> 이미지 데이터 (프레임 단위로 자르기)

```python
import cv2
import os

print(cv2.__version__)
```

```python
# 경로 중에 한글이 있으면 cv2.imwrite 안됌!
root_path = 'D://Weather conditions driving classification data (YouTube)//'

file_paths = [root_path+'original_video//Normal.mp4',root_path+'original_video//Rainy.mp4',root_path+'original_video//Snowy.mp4']

for file_path in file_paths:
    video = cv2.VideoCapture(file_path) #'' 사이에 사용할 비디오 파일의 경로 및 이름을 넣어주도록 함
    # if not video.isOpened():
    #     print("Could not Open :", file_path)
    #     exit(0)
    length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)

    print(file_path)
    print("length :", length)
    print("width :", width)
    print("height :", height)
    print("fps :", fps)

    try:
        if not os.path.exists(root_path+'//images//'+file_path.split('//')[-1][:-4]):
            os.makedirs(root_path+'//images//'+file_path.split('//')[-1][:-4])
            print(root_path+'//images//'+file_path.split('//')[-1][:-4])
    except OSError:
        print ('Error: Creating directory. ' +  root_path+'//images//'+file_path.split('//')[-1][:-4])

    count = 0

    while(video.isOpened()):
        
        # Normal, Rainy, Snowy 3개 클래스 각 5000개씩(영상 시간 고려)
        ret, image = video.read()

        # if(int(video.get(1)) % fps == 0): #앞서 불러온 fps 값을 사용하여 1초마다 추출
        if(int(video.get(1)) % 20 == 0): # 5 프레임당 1개씩 추출
            cv2.imwrite(root_path+'images//'+file_path.split('//')[-1][:-4] + "//frame%d.jpg" % count, image)
            print(root_path+'images//'+file_path.split('//')[-1][:-4] + "//frame%d.jpg" % count)
            # print('Saved frame number :', str(int(video.get(1))))
            count += 1

        if count > 5000:
            break
            
    video.release()

```

### ※Youtube Snowy 클래스 데이터 추가 (AI HUBd의 Snwoy 데이터는 노면 상태가 Rainy와 유사)

 * Snowy 데이터만 추가

* AI hub Snowy 경로로 추가

```python
# 경로 중에 한글이 있으면 cv2.imwrite 안됌!
youtube_root_path = 'D://Weather conditions driving classification data (youtube)//'
detail_path = 'original_video//additional//'

ai_hub_root_path = 'D://Weather conditions driving classification data (AI hub)//'


file_paths = [youtube_root_path+detail_path+'Snowy2.mp4' ,youtube_root_path+detail_path+'Snowy3.mp4',youtube_root_path+detail_path+'Snowy4.mp4']

total_count = 0

for file_path in file_paths:
    video = cv2.VideoCapture(file_path) #'' 사이에 사용할 비디오 파일의 경로 및 이름을 넣어주도록 함
    # if not video.isOpened():
    #     print("Could not Open :", file_path)
    #     exit(0)
    length = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = video.get(cv2.CAP_PROP_FPS)

    print(file_path)
    print("length :", length)
    print("width :", width)
    print("height :", height)
    print("fps :", fps)

    try:
        if not os.path.exists(ai_hub_root_path+'images//Snowy//'):
            os.makedirs(ai_hub_root_path+'images//Snowy//')
            print(ai_hub_root_path+'images//Snowy//')
    except OSError:
        print ('Error: Creating directory. ' +  ai_hub_root_path+'images//Snowy//')

    count = 0
    while(video.isOpened()):
        
        # Normal, Rainy, Snowy 3개 클래스 각 5000개씩(영상 시간 고려)
        ret, image = video.read()

        # if(int(video.get(1)) % fps == 0): #앞서 불러온 fps 값을 사용하여 1초마다 추출
        if(int(video.get(1)) % 20 == 0): # 5 프레임당 1개씩 추출
            cv2.imwrite(ai_hub_root_path+'images//Snowy//' + "frame%d.jpg" % total_count, image)
            print(ai_hub_root_path+'images//Snowy//'+ "frame%d.jpg" % total_count)

            count += 1
            total_count += 1
        
        if total_count > 5000:
            break

        if count > length//20-1:
            break
            
    video.release()
```
