```python
from glob import glob
baseFolder = 'C:/05_31_img/'
fileList = glob(baseFolder+'*')
fileList
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm

sh3 , sh4 = [],[]

for i in tqdm(fileList):
    img = Image.open(i).resize((100,200))

    # /255 스케일링 안해도 plt.imshow에는 이상없음
    imgArr = np.array(img)/255
    # imgArr = np.array(img)

    # shape 출력
    print('file :',i,imgArr.shape)
    # or 헤더 파일로 출력해볼 수도 있음

    if imgArr.shape[2] == 3:
        sh3.append(i)
    else:
        sh4.append(i)

    print("원본------")
    plt.imshow(imgArr)
    plt.show()
    print("R------")
    plt.imshow(imgArr[:,:,0])
    plt.show()
    print("G------")
    plt.imshow(imgArr[:,:,1])
    plt.show()
    print("B------")
    plt.imshow(imgArr[:,:,2])
    plt.show()
    print("B < 0.5 to 0 ------")
    imgArr[imgArr[:,:,2]<0.5] = 0
    plt.imshow(imgArr)
    plt.show()
```
